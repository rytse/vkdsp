```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pandas
```

```python
ts = np.linspace(0, 10, 1000)
ys = np.exp(1000.0j * ts) + np.exp(2.0j*ts)
plt.plot(ts, np.abs(ys))
plt.show()

ys_spec = np.fft.fft(ys)
plt.plot(ts, np.abs(ys_spec))
plt.show()
```

```python
ys_spec.tofile("ys.bin")
```
